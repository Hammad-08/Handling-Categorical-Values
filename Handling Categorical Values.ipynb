## Ordinal Encoding

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.drop(columns=["purchased"]),df["purchased"],random_state=42,test_size=0.3)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
```

```python
X_train
```
